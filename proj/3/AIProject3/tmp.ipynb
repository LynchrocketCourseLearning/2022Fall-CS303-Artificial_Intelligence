```python
import torch
data = torch.load("data.pth")
label = data["label"]
feature = data["feature"]
```

```python
from my_network import MyClassifier

net = MyClassifier()
optimizer = torch.optim.SGD(net.parameters(), lr=0.02)
loss_func = torch.nn.CrossEntropyLoss()
```

```python
myModel = torch.load('myModel.pth', map_location='cpu')
```

```python
myModel
```

```python
net
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import accuracy_score
import torch.nn.functional as F
```

```python
X_train, X_test, y_train, y_test = tts(feature, label, test_size=0.2, random_state=0)
```

```python
for x, y in zip(X_train, y_train):
    pred = net(x)
    loss = loss_func(pred, y)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
pred_list = []
for x, y in zip(X_test, y_test):
    prediction = torch.nn.Softmax(dim=0)(myModel(x))
    y_pred = prediction.argmax(0)
    pred_list.append(y_pred)
acc = accuracy_score(y_pred=pred_list, y_true=y_test.data.numpy())
print(acc)
```
